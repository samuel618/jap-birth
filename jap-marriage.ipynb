# Data Cleaning

```python
clean_df = selected_df.copy()
# Sort by year
clean_df = clean_df.sort_values("year").reset_index(drop=True)

# Remove duplicate rows
clean_df = clean_df.drop_duplicates()

print("Shape after cleaning duplicates:", clean_df.shape)
clean_df.head()
```

```python
# Convert all columns except year to numeric
for col in clean_df.columns:
    if col != "year":
        clean_df[col] = pd.to_numeric(clean_df[col], errors="coerce")

clean_df.info()
```

```python
# Columns where husband is Japanese and wife is foreign
husb_foreign_cols = [col for col in clean_df.columns if col.startswith("husb_jpn_wife_")]

# Columns where wife is Japanese and husband is foreign
wife_foreign_cols = [col for col in clean_df.columns if col.startswith("wife_jpn_husb_")]

print("Husband Japanese + foreign wife columns:")
print(husb_foreign_cols)

print("\nWife Japanese + foreign husband columns:")
print(wife_foreign_cols)
```

```python
# Combine all foreign-country categories
clean_df["total_husb_jpn_foreign_wife"] = clean_df[husb_foreign_cols].sum(axis=1, skipna=True)
clean_df["total_wife_jpn_foreign_husband"] = clean_df[wife_foreign_cols].sum(axis=1, skipna=True)

# Drop 1965 to keep a continuous yearly sequence from 1970 onward
clean_df = clean_df[clean_df["year"] != 1965].reset_index(drop=True)

# Total marriages involving a foreign spouse
clean_df["total_foreign_spouse"] = (
    clean_df["total_husb_jpn_foreign_wife"] +
    clean_df["total_wife_jpn_foreign_husband"]
)

# Total marriages
clean_df["total_marriages"] = (
    clean_df["both_japanese"] +
    clean_df["total_foreign_spouse"]
)
clean_df.head()
```

# Data Preparation

```python
jap_copy_df = clean_df[
    [
        "year",
        "both_japanese",
        "total_husb_jpn_foreign_wife",
        "total_wife_jpn_foreign_husband",
        "total_foreign_spouse",
        "total_marriages"
    ]
].copy()

jap_copy_df.head()
```

```python
jap_df = jap_copy_df.melt(
    id_vars="year",
    var_name="type",
    value_name="amount"
)

jap_df.head()
```
